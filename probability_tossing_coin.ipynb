```python
# import packages
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
# toss a coin
def toss():
  outcome = random.randint(0,1)
  if outcome == 0:
    return 'H'
  return 'T'
```

```python
# tossing coin 10000 times
data = [toss() for _ in range(10000)]
```

```python
_h,_t,time = [],[],[]
for i in range(20,10000,10):
  _h.append(data[:i].count('H')/i)
  _t.append(data[:i].count('T')/i)
  time.append(i)
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(time,_h,c='blue')
ax.plot(time,_t,c='green')
ax.plot(time, [.50 for _ in range(20,10000,10)], c='red')
plt.xlabel("time →")
plt.ylabel("probability →")
plt.show()
```
